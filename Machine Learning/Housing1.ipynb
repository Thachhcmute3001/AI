```python
#Machine learning
import pandas as pd

url = 'housing1.csv'
dataframe = pd.read_csv(url)
```

```python
data = dataframe.drop(['ocean_proximity'], axis = 'columns')
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'longitude', y = 'latitude',figsize = (9,9), color = 'black')
```

```python
#Mo hinh bac 2
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['longitude'],data['longitude']**2]).T, y = data['latitude'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame([data['longitude'],data['longitude']**2]).T, y = data['latitude'])*100 #Accuracy
```

```python
import numpy as np
range = np.arange(-124,-114,1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'longitude', y = 'latitude', figsize = (9,9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```
