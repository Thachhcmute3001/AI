```python
#Machine learning
import pandas as pd

url = 'Salary_Data.csv'
data = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
''' mpg: mile per galon
    disp: the tich buon dot
    np: ma luc
    wt: khoi luong  '''
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'Salary',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])*100 #Accuracy
```

```python
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'Salary', figsize = (9,9), color = 'black', xlim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'], y_pred, color = 'red')
```

```python
sns.pairplot(data)
```

```python
plt.plot(data.YearsExperience,data.Salary,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
model = LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience'])).round(1)
calculation = pd.DataFrame(np.c_[data.Salary, y_pred], columns = ['Original Salary','Predicted Salary'])
print(calculation)
```
