```python
#Machine learning
import pandas as pd

url = 'Advertising.csv'
data = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
''' saleses: mile per galon
    disp: the tich buon dot
    np: ma luc
    TV: khoi luong  '''
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'sales', y = 'TV',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['TV'],data['TV']**2]).T, y = data['sales'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame([data['TV'],data['TV']**2]).T, y = data['sales'])*100 #Accuracy
```

```python
import numpy as np
range = np.arange(1.5,300,0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```

```python
#Mo hinh bac 5
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5]).T, y = data['sales'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5]).T, y = data['sales'])*100 #Accuracy
```

```python
range = np.arange(1.5,300,0.1)
predictor = pd.DataFrame([range, range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```
