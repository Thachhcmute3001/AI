```python
#Machine learning
import pandas as pd

url = 'passenger1.csv'
dataframe = pd.read_csv(url)
```

```python
data = dataframe.drop(['Passenger'], axis = 'columns')
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)',figsize = (9,9), color = 'black')
```

```python
#Mo hinh bac 2
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2]).T, y = data['Dollars (Y)'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2]).T, y = data['Dollars (Y)'])*100 #Accuracy
```

```python
import numpy as np
range = np.arange(0,5000,10)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```
