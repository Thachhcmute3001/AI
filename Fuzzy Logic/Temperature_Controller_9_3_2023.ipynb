```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
#input
universe = np.linspace(18,30, 100) 
universe1 = np.linspace(-5,60, 100) #%
#output
universe2 = np.linspace(0,255,100)


setpoint = ctrl.Antecedent(universe, 'setpoint')
sensor = ctrl.Antecedent(universe1, 'sensor')
pwm = ctrl.Consequent(universe2, 'pwm')

#input
setpoint['cold'] = fuzzy.gaussmf(setpoint.universe,18,7)
setpoint['cool'] = fuzzy.gaussmf(setpoint.universe,25,7)
setpoint['warm'] = fuzzy.gaussmf(setpoint.universe,28,7)
setpoint.view()

sensor['low'] = fuzzy.gaussmf(sensor.universe,-5,12)
sensor['medium'] = fuzzy.gaussmf(sensor.universe,25,12)
sensor['high'] = fuzzy.gaussmf(sensor.universe,45,12)
sensor.view()
#output
pwm['low'] = fuzzy.gaussmf(pwm.universe,50,70)
pwm['medium'] = fuzzy.gaussmf(pwm.universe,160,70)
pwm['high'] = fuzzy.gaussmf(pwm.universe,210,70)
pwm.view()
#rule
rule_0 = ctrl.Rule(antecedent = ((setpoint['cold'] & sensor['low']) | (setpoint['cool'] & sensor['low']) | (setpoint['warm'] & sensor['medium'])), consequent = pwm['low'])
rule_1 = ctrl.Rule(antecedent = ((setpoint['cold'] & sensor['medium']) | (setpoint['cool'] & sensor['medium']) | (setpoint['warm'] & sensor['high']) | (setpoint['warm'] & sensor['low'])), consequent = pwm['medium'])
rule_2 = ctrl.Rule(antecedent = ((setpoint['cold'] & sensor['high']) | (setpoint['cool'] & sensor['high'])), consequent = pwm['high'])


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

System  = ctrl.ControlSystem(rules = [rule_0, rule_1, rule_2])
sim = ctrl.ControlSystemSimulation(System, flush_after_run= 61*61 + 1)

upSampled1 = np.linspace(18,30, 61)
upSampled2 = np.linspace(-5,60, 61)
x,y = np.meshgrid(upSampled1, upSampled2)
z = np.zeros_like(x)

for i in range(61):
  for j in range(61):
    sim.input['setpoint']= x[i,j]
    sim.input['sensor'] = y[i,j]
    sim.compute()
    z[i,j] = sim.output['pwm']

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection = '3d')
suf = ax.plot_surface(x,y,z, rstride = 1, cstride = 1, cmap = 'viridis',
                      linewidth= 0.4)
cset = ax.contourf(x,y,z, zdir= 'z', offset = 0, cmap= 'viridis', anpha= 0.5)
cset = ax.contourf(x,y,z, zdir= 'x', offset = 30, cmap= 'viridis', anpha= 0.5)
cset = ax.contourf(x,y,z, zdir= 'y', offset = 60, cmap= 'viridis', anpha= 0.5)
ax.view_init(30,200)



     
```
