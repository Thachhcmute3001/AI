```python
#Machine learning
import pandas as pd
url = 'mtcars.csv'
data = pd.read_csv(url)

```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
''' mpg: mile per galon
    disp: the tich buon dot
    np: ma luc
    wt: khoi luong  '''
```

```python
correlation = data.corr()
print(correlation)
#Gia tri abs(x) voi x là biến, với giá trị abs(x) ngoại trừ chính nó
#thì giá trị lớn nhất trong các biến còn lại sẽ ảnh hưởng lớn nhất đến biến đang xét
#ví dụ: xét mpg, thì giá trị abs(wt) = 0.867 là có ảnh hưởng lớn nhất
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame(data['wt']), y = data['mpg'])*100 #Accuracy
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_pred, color = 'red')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['wt'],data['wt']**2]).T, y = data['mpg'])
print(model.intercept_)# y = ax + b --> b = intercept
print(model.coef_)#y = ax + b --> a = cof
model.score(X = pd.DataFrame([data['wt'],data['wt']**2]).T, y = data['mpg'])*100 #Accuracy
```

```python
import numpy as np
range = np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0,7))
plt.plot(range, y_pred, color = 'red')
```
