```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

food = ctrl.Antecedent(np.arange(0,11,1), 'Food')
service = ctrl.Antecedent(np.arange(0,11,1), 'Service')
tip = ctrl.Consequent(np.arange(10,31,0.05), 'Tip')
#------------------#
food['poor'] = fuzz.trimf(food.universe,[0, 0, 5])
food['average'] = fuzz.trimf(food.universe, [0, 5, 10])
food['good'] = fuzz.trimf(food.universe, [5, 10, 10])
#------------------#
service['poor'] = fuzz.trimf(service.universe,[0, 0, 5])
service['average'] = fuzz.trimf(service.universe, [0, 5, 10])
service['good'] = fuzz.trimf(service.universe, [5, 10, 10])
#------------------#
tip['less'] = fuzz.trimf(tip.universe,[10, 10, 20])
tip['normal'] = fuzz.trimf(tip.universe, [10, 20, 30])
tip['much'] = fuzz.trimf(tip.universe, [20, 30, 30])

rule_1 = ctrl.Rule(food['poor'] & service['poor'],tip['less'])
rule_2 = ctrl.Rule(food['poor'] & service['average'], tip['less'])
rule_3 = ctrl.Rule(food['poor'] & service['good'], tip['less'])
rule_4 = ctrl.Rule(food['average'] & service['poor'], tip['less'])
rule_5 = ctrl.Rule(food['average'] & service['average'],tip['normal'])
rule_6 = ctrl.Rule(food['average'] & service['good'],tip['much'])
rule_7 = ctrl.Rule(food['good'] & service['poor'],tip['less'])
rule_8 = ctrl.Rule(food['good'] & service['average'],tip['normal'])
rule_9 = ctrl.Rule(food['good'] & service['good'],tip['much'])

rule = [rule_1, rule_2, rule_3, rule_4, rule_5, rule_6, rule_7, rule_8, rule_9]
tipping_ctrl = ctrl.ControlSystem(rule)
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['Food'] = 10
tipping.input['Service'] = 10
tipping.compute()
print(tipping.output['Tip'])
tip.view(sim = tipping)
#food.view()
#service.view()
#tip.view()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

time = ctrl.Antecedent(np.arange(5,180,10), 'time')
rice = ctrl.Antecedent(np.arange(0, 1000, 100), 'rice')
power = ctrl.Consequent(np.arange(10, 100, 1), 'power')
#------------------#
time['very little'] = fuzz.trimf(time.universe,[5, 10, 20])
time['little'] = fuzz.trimf(time.universe, [10, 15, 20])
time['average'] = fuzz.trimf(time.universe, [20, 25, 30])
time['much'] = fuzz.trimf(time.universe, [30, 60, 180])
time['so much'] = fuzz.trimf(time.universe, [60, 180, 180])
#------------------#
rice['very little'] = fuzz.trimf(rice.universe,[0, 100, 200])
rice['little'] = fuzz.trimf(rice.universe, [100, 300, 450])
rice['average'] = fuzz.trimf(rice.universe, [300, 600, 700])
rice['much'] = fuzz.trimf(rice.universe, [600, 800, 1000])
rice['so much'] = fuzz.trimf(rice.universe, [800, 850, 1000])
#------------------#
power['very low'] = fuzz.trimf(power.universe,[10, 15, 20])
power['low'] = fuzz.trimf(power.universe, [15, 30, 40])
power['middle'] = fuzz.trimf(power.universe, [30, 60, 70])
power['much'] = fuzz.trimf(power.universe, [60, 80, 85])
power['so much'] = fuzz.trimf(power.universe, [80, 100, 100])

rule_1 = ctrl.Rule(time['very little'] & rice['very little'], power['very low'])
rule_2 = ctrl.Rule(time['very little'] & rice['little'], power['very low'])
rule_3 = ctrl.Rule(time['very little'] & rice['average'], power['very low'])
rule_4 = ctrl.Rule(time['very little'] & rice['much'], power['very low'])
rule_5 = ctrl.Rule(time['very little'] & rice['so much'], power['very low'])

rule_6 = ctrl.Rule(time['little'] & rice['very little'], power['low'])
rule_7 = ctrl.Rule(time['little'] & rice['little'], power['low'])
rule_8 = ctrl.Rule(time['little'] & rice['average'], power['low'])
rule_9 = ctrl.Rule(time['little'] & rice['much'], power['low'])
rule_10 = ctrl.Rule(time['little'] & rice['so much'], power['low'])

rule_11 = ctrl.Rule(time['average'] & rice['very little'], power['middle'])
rule_12 = ctrl.Rule(time['average'] & rice['little'], power['middle'])
rule_13 = ctrl.Rule(time['average'] & rice['average'], power['middle'])
rule_14 = ctrl.Rule(time['average'] & rice['much'], power['middle'])
rule_15 = ctrl.Rule(time['average'] & rice['so much'], power['middle'])

rule_16 = ctrl.Rule(time['much'] & rice['very little'], power['much'])
rule_17 = ctrl.Rule(time['much'] & rice['little'], power['much'])
rule_18 = ctrl.Rule(time['much'] & rice['average'], power['much'])
rule_19 = ctrl.Rule(time['much'] & rice['much'], power['much'])
rule_20 = ctrl.Rule(time['much'] & rice['so much'], power['much'])

rule_21 = ctrl.Rule(time['so much'] & rice['very little'], power['so much'])
rule_22 = ctrl.Rule(time['so much'] & rice['little'], power['so much'])
rule_23 = ctrl.Rule(time['so much'] & rice['average'], power['so much'])
rule_24 = ctrl.Rule(time['so much'] & rice['much'], power['so much'])
rule_25 = ctrl.Rule(time['so much'] & rice['so much'], power['so much'])



rule = [rule_1, rule_2, rule_3, rule_4, rule_5, rule_6, rule_7, rule_8, rule_9, rule_10, rule_11, rule_12, rule_13, rule_14, rule_15, rule_16, rule_17, rule_18, rule_19, rule_20, rule_21, rule_22, rule_23, rule_24, rule_25]
tipping_ctrl = ctrl.ControlSystem(rule)
powering = ctrl.ControlSystemSimulation(tipping_ctrl)
powering.input['time'] = 180
powering.input['rice'] = 1000
powering.compute()
print(powering.output['power'])
power.view(sim = powering)
#food.view()
#service.view()
#tip.view()
```
